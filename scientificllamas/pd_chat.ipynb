```python
import sys
import PyPDF2
from PyQt5.QtWidgets import *
from PyQt5.QtCore import *
from PyQt5.QtGui import *
from PyQt5.QtMultimedia import *
from PyQt5.QtMultimediaWidgets import *

class pd_gpt(QWidget):
    video_playlist = QMediaPlaylist()

    def __init__(self):
        super().__init__()

        # Set up the GUI
        self.initUI()

        self.initMusic()

    def initUI(self):
        # Create a button to choose a PDF file
        btn_choose = QPushButton("Choose PDF", self)
        btn_choose.setToolTip("Choose PDF")
        btn_choose.setSizePolicy(QSizePolicy.Fixed, QSizePolicy.Fixed)
        btn_choose.clicked.connect(self.choose_pdf)

        # Create a button to extract text
        btn_extract = QPushButton("Extract Text", self)
        btn_extract.setToolTip("Extract Text")
        btn_extract.setSizePolicy(QSizePolicy.Fixed, QSizePolicy.Fixed)
        btn_extract.clicked.connect(self.extract_text)

        # Create a text box to display the extracted text
        self.text_box = QTextEdit(self)

        # Create a video widget to display the looping video
        self.video_widget = QVideoWidget(self)
        self.video_widget.setFixedSize(200, 150)
        self.video_widget.setMinimumSize(1, 1)

        # Create a media player object to play the looping video
        self.video_player = QMediaPlayer()
        self.video_player.setMedia(QMediaContent(QUrl.fromLocalFile(r"C:\Users\gzeba\Downloads\IndolentDirectIvorybackedwoodswallow-mobile.mp4")))
        self.video_player.setVideoOutput(self.video_widget)
        self.video_player.setVolume(0)

        # Add the looping video to the playlist
        self.video_playlist.addMedia(QMediaContent(QUrl.fromLocalFile(r"C:\Users\gzeba\Downloads\IndolentDirectIvorybackedwoodswallow-mobile.mp4")))

        # Set the playlist to loop
        self.video_playlist.setPlaybackMode(QMediaPlaylist.Loop)

        # Set the playlist to the media player
        self.video_player.setPlaylist(pd_gpt.video_playlist)

        # Start playing the video
        self.video_player.play()

        # Create a play button
        self.btn_play = QPushButton(self)
        self.btn_play.setToolTip("Play")
        self.btn_play.setIcon(self.style().standardIcon(QStyle.SP_MediaPlay))
        self.btn_play.clicked.connect(self.play_music)

        # Create a stop button
        self.btn_stop = QPushButton(self)
        self.btn_stop.setToolTip("Stop")
        self.btn_stop.setIcon(self.style().standardIcon(QStyle.SP_MediaStop))
        self.btn_stop.clicked.connect(self.stop_music)

        # Create a mute button
        self.btn_mute = QPushButton(self)
        self.btn_mute.setToolTip("Mute")
        self.btn_mute.setIcon(self.style().standardIcon(QStyle.SP_MediaVolumeMuted))
        self.btn_mute.clicked.connect(self.toggle_mute)

        # Create a label to indicate the mute state
        self.lbl_mute = QLabel(self)
        self.lbl_mute.setPixmap(self.style().standardPixmap(QStyle.SP_MediaVolumeMuted).scaled(20, 20))
        self.lbl_mute.setAlignment(Qt.AlignCenter)

        # Create a horizontal layout for the play, stop, and mute buttons
        hbox = QHBoxLayout()
        hbox.addStretch(1)
        hbox.addWidget(self.btn_play)
        hbox.addWidget(self.btn_stop)
        hbox.addWidget(self.btn_mute)
        hbox.addWidget(self.lbl_mute)
        hbox.addStretch(1)

        # Create a vertical layout for the buttons and text box
        vbox = QVBoxLayout()
        vbox.addWidget(btn_choose)
        vbox.addWidget(btn_extract)
        vbox.addWidget(self.text_box)
        vbox.addLayout(hbox)
        
        # Create a label to show the logo
        logo_label = QLabel(self)
        logo_pixmap = QPixmap(r'C:\Users\gzeba\Downloads\pdfchat.jpg')
        logo_pixmap = logo_pixmap.scaledToHeight(125, Qt.SmoothTransformation)
        logo_pixmap = logo_pixmap.scaled(logo_pixmap.width(), logo_pixmap.height(), Qt.KeepAspectRatio)
        logo_label.setPixmap(logo_pixmap)
        logo_label.setAlignment(Qt.AlignHCenter)

        # Create a vertical layout for the logo and the main layout
        main_layout = QVBoxLayout()
        main_layout.addWidget(logo_label)
        main_layout.addLayout(vbox)

        # Set the main layout
        self.setLayout(main_layout)

        # Create a layout for the video widget and add it to the main layout
        video_layout = QHBoxLayout()
        video_layout.addWidget(self.video_widget)
        vbox.addLayout(video_layout)

        # Initialize the mute state
        self.is_muted = False

        # Set up the window
        self.setLayout(vbox)
        self.setGeometry(100, 100, 640, 550)
        self.setWindowTitle("pd.ChAT!")
        self.show()

    def initMusic(self):
        # Create a media player object to play the music
        self.music_player = QMediaPlayer()
        self.music_playlist = QMediaPlaylist()
        file_path = r"C:\\Users\\gzeba\\Downloads\\Boys (Summertime Love) - Sabrina (Salerno) - backingtrackx.com.mp3"
        self.music_playlist.addMedia(QMediaContent(QUrl.fromLocalFile(file_path)))
        self.music_playlist.setPlaybackMode(QMediaPlaylist.Loop)
        self.music_player.setPlaylist(self.music_playlist)
        self.music_player.setVolume(50)
        self.music_player.play()

    def choose_pdf(self):
        # Let the user choose a PDF file
        file_path, _ = QFileDialog.getOpenFileName(
            self,
            "Choose a PDF file",
            "",
            "PDF Files (*.pdf)"
        )

        # Store the file path
        self.file_path = file_path

    def extract_text(self):
        # Open the PDF file in binary mode
        with open(self.file_path, "rb") as pdf_file:
            # Create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(pdf_file)

            # Clear the text box
            self.text_box.clear()

            # Loop through each page in the PDF file
            for page_num in range(len(pdf_reader.pages)):
                # Get the page object
                page = pdf_reader.pages[page_num]

                # Extract the text from the page
                text = page.extract_text()

                # Append the text to the text box
                self.text_box.append(text)

    def play_music(self):
        # Start playing the music
        self.music_player.play()

    def stop_music(self):
        # Stop playing the music
        self.music_player.stop()

    def toggle_mute(self):
        if self.is_muted:
            self.is_muted = False
            self.music_player.setMuted(False)
            self.btn_mute.setIcon(self.style().standardIcon(QStyle.SP_MediaVolume))
            self.lbl_mute.setPixmap(self.style().standardPixmap(QStyle.SP_MediaVolume).scaled(20, 20))
        else:
            self.is_muted = True
            self.music_player.setMuted(True)
            self.btn_mute.setIcon(self.style().standardIcon(QStyle.SP_MediaVolumeMuted))
            self.lbl_mute.setPixmap(self.style().standardPixmap(QStyle.SP_MediaVolumeMuted).scaled(20, 20))
    
    def closeEvent(self, event):
        self.music_player.stop()
        event.accept()

if __name__ == "__main__":
    # Create the application
    app = QApplication(sys.argv)

    app.setStyle("fusion")
    palette = QPalette()
    palette.setColor(QPalette.Window, QColor(53, 53, 53))
    palette.setColor(QPalette.WindowText, Qt.white)
    palette.setColor(QPalette.Base, QColor(25, 25, 25))
    palette.setColor(QPalette.AlternateBase, QColor(53, 53, 53))
    palette.setColor(QPalette.ToolTipBase, Qt.white)
    palette.setColor(QPalette.ToolTipText, Qt.white)
    palette.setColor(QPalette.Text, Qt.white)
    palette.setColor(QPalette.Button, QColor(53, 53, 53))
    palette.setColor(QPalette.ButtonText, Qt.white)
    palette.setColor(QPalette.BrightText, Qt.red)
    palette.setColor(QPalette.Link, QColor(42, 130, 218))
    palette.setColor(QPalette.Highlight, QColor(42, 130, 218))
    palette.setColor(QPalette.HighlightedText, Qt.black)
    app.setPalette(palette)

    # Create the PDFExtractor window
    window = pd_gpt()

    # Run the application
    sys.exit(app.exec_())

    # Create the main window
    #window = MainWindow()
    #window.show()
```
